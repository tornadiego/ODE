## <h2><b>Second order oridinary differential equation</b>



Any differential equation(DE) is a relation (every equation is a relation) between a function and its derivatives When the derivatives in the equation is of one variable only, it's called <i>oridinary</i> (but the function itself could be multivariabled). For instance
\begin{equation}
\frac{dy}{dt}=y(x,t) \\
7\frac{d^2y}{dt^2}+69\frac{dy}{dt}=y^2+e^x
\end{equation}
The derivative depends on the function $y$ which in turn depends on the independent variable $x$. Keeping this in mind we can always represent <i>second order differential equation as coupled first order differential equation.</i>

# <center> <h1><b> EULER'S METHOD </b>    </center>

---


Let $y(t)$ be a function of variable $t$, according to taylor's expansion
then the function at $t + \triangle t$ would be equal to 

<center> $y(t+\triangle t) = y(t) + \triangle t \;.\;y'(y(t),\; t)+\bigcirc (\triangle t^2)  $</center> 

where $\omicron$ shows the order of approximation we are undertaking. All other derivatives will be omitted from the expression

#<h1><center><b>RUNGE-KUTTA METHOD</b>

---

Runge-Kutta provides you highly meticluously woven expressions to solve LDEs  
<center>$y(t+\triangle t) = y(t) + \triangle t \; .\, y'(y(t) + \frac{k(t)}{2}, t+\frac{\triangle t}{2}) + \bigcirc (\triangle t^3)$

where

$k(t)=\triangle t . \, y'(y,t)$
</center> 

other symbols have usual meaning. This expression (called, RK-2) says that value of function at some interval later is sum of function itself and increment when the slope is calculated at the mid-point of the interval.

The problem with euler's technique is evidently the order of approximation it undertakes.
So Range-kutta used higher order derivative to calculate function after successive intervals. 

## ***Origin of expression***

From taylor's expansion

$y(t + \triangle t) y(t) + \triangle t\, .\, y'(y,t) + \frac{1}{2}\triangle t^2 y''(y', \,y,\, t) + \bigcirc (\triangle t^3), \tag{1}$

We will omit error bar hitherto
<br>

<center>$y(t + \triangle t) = y(t) + \triangle t\, .[\, y'(t) + \frac{1}{2}\triangle t \, y''] \tag{2} $</center>

The term $ (y'(t) + \frac{1}{2}\triangle t \: y''(y,t))$ can be thought as the slope of the function at half interval later.   *(Note now the function is taking more than one arguments)*

i.e.

Let,  $g(t) = y'(t) $

$\implies g(t+ \frac{1}{2} \triangle t) = g(t) + \frac{1}{2} \triangle t \, g'(t)  $
which is exactly the expression as above. Therefore, the second equation can be written as 

<center>$y(t + \triangle t) = y(t) + \triangle t\, .y'(y + \frac{1}{2}\triangle t. y',\:t +  \frac{1}{2}\triangle t)  \tag{3} $</center>

The equation (3) is often called RK-2, as we have another Range-kutte formula which uses 5th order of approximation and so is called RK-4. ***RK-4*** uses four intermediate slopes to carry out the increment in the function.<br><br>


\begin{equation}
y(t+\triangle t) = y(t) + \frac{1}{6} [k_1+2k_2+2k_3+k_4] + \bigcirc (\triangle t^5) \\
where \\ 
\\
k_1 = \triangle t.\;y'(y,t) \\
k_2 = \triangle t.\;y'(y + \frac{1}{2} k_1,t + \frac{1}{2} \triangle t) \\
k_3 = \triangle t.\;y'(y + \frac{1}{2} k_2,\:t + \frac{1}{2} \triangle t)) \\
k_4 = \triangle t.\;y'(y +  k_3, \; t + \triangle t) 
\end{equation}

# <center><b><h1>VERLET'S METHOD </b></center>

---

Now we will be talking about methods which are mystifying in the way that they give suprising accuracies in certain cases. We start by describing the Verlet's algorithm. 

Let a particle in a system having parameters given, i.e. its velocity $v$ and position $p$ is given at some time $t_0$. The particle is in the conservative field, therefore acceleration of the particle is only dependent on position of the particle ($a = a(p))$.  Then we can use following method to find its further configuration:
\begin{align}
p(t + \frac{1}{2} \triangle t) &= p(t) + \frac{1}{2} \triangle t. \; v(t) \tag{1}, \\
v(t + \triangle t) &= v(t) + \triangle t. \; a(\,p( t)\,) \tag{2}  \\
p(t + \triangle t) &= p(t + \frac{1}{2} \triangle t) + \frac{1}{2} \triangle t. \; v(t + \triangle t) \tag{3}\\
\end{align}

Specifically, this set of equations is called position-verlet. Symmetrically you can also obtain velocity-verlet. This method gives no error in *energy* values, even when it is using the third order approximation, for the reasons as follow.

## Vector equations

If the problem is of more than one dimensional, that is now position is defined by $(x,y,z,..)$ cordinates and so will the velocity $(v_x,v_y,v_z,..)$. We can split equations based on components as either component are independent. 

#TIPS FOR WRITING THE CODE

We will show how to write the problems with providing you a model problem of *damped harmonic oscillator*. For instance,

<tt> Consider a bob of mass '$m$' is weld to a very light rod of lenght $l$, which in turn is fitted to a wall such that it can pivot about one end (opposite end to the bob). The whole set-up is submerged in a viscous liquid on earth such that it provides the drag to the object which is proportional to the object's velocity ($b$ being the proportionality constant). Applyting Newton's law we get the differential equation for this situation as following:
</tt>

\begin{equation}
m\frac{d^2\overrightarrow{P}}{dt^2} -b\frac{d\overrightarrow{P}}{dt}-\omega^2\overrightarrow{P}=0 
\end{equation}

> where

>>>>    $w= \sqrt{\frac{g}{l}}$ 

>>>>   $ T_0 = \frac{1}{\omega} $

>>>>    $ \overrightarrow{P} = x \, \hat{i} + y \, \hat{j} $

>>>>    $ \overrightarrow{V} = {d \overrightarrow{P}}/{dt} =  dx/dt \, \hat{i} + dy/dt \, \hat{j} $

>>>>    $ \overrightarrow{A} = {d^2 \overrightarrow{P}}/{dt^2} =  d^2x/dt^2 \, \hat{i} + d^2y/dt^2 \, \hat{j} $


```
#####importing libraries

from matplotlib import pylab as plt
import math as m

#####Defining velocity as a function
def Vel(X, Y, T):
    return Y

#####Defining acceleration as a function
def acc(X, Y, T):
    return -(4.0 * (m.pi * m.pi) * m.sin(X)) - (b * Y) + (A * m.sin(k * T))

```

```
# initial conditions and constants of eqn
w = 2. / 3.
T_O = 2 * m.pi
q = 0.5
F_O = 1.12

X = 0.0
Y = 0.0
T = 0.0

T_end = 1000
N = 10000
del_T = T_end / float(N)

b = q * T_O
A = F_O * T_O * T_O
k = w * T_O

# for plotting
T_list = []
X_list = []
Y_list = []
```

```
# to sort out only limited numbers
n = 50  # for number of multiple

for x in range(0, N):

    # mediate points slopes for position and velocity

    K1 = del_T * acc(X, Y, T)
    K5 = del_T * Vel(X, Y, T)
    K2 = del_T * acc(X + Y * del_T / 2, Y + K1 / 2, T + del_T / 2)
    K6 = del_T * Vel(X + Y * del_T / 2, Y + K5 / 2, T + del_T / 2)
    K3 = del_T * acc(X + Y * del_T / 2, Y + K2 / 2, T + del_T / 2)
    K7 = del_T * Vel(X + Y * del_T / 2, Y + K6 / 2, T + del_T / 2)
    K4 = del_T * acc(X + del_T * Y / 2, Y + K3 / 2, T + del_T / 2)
    K8 = del_T * Vel(X + del_T * Y / 2, Y + K7 / 2, T + del_T / 2)
    X_new = X + (1 / 6) * (K5 + 2 * K6 + 2 * K7 + K8)
    Y_new = Y + (1 / 6) * (K1 + 2 * K2 + 2 * K3 + K4)

    if(X_new<=m.pi and X_new>=-m.pi):
        X_new=X_new
    if (X_new > m.pi):
        X_new = X_new - 2 * m.pi
    elif (X_new < -m.pi):
        X_new = 2 * m.pi + X_new

    X = X_new
    Y = Y_new
    T = T + del_T

#
    if (T < T_end /1.1):# and del_T / 2 > m.fabs(T - n / w)):
        n = n + 1
    X_list.append(X)
    Y_list.append(Y)
    T_list.append(T)

plt.plot(X_list, Y_list, '.', color="black", markersize="10")
# plt.xlim(-4,4)
# plt.ylim(-15,15)
# plt.plot(T_list, X_list,'.',color="red")
plt.show()  # 


```
